```python
from math import *
class Level:
    def __init__(self, value):
        self.value = value

    def from_rms(self, rms):
        self.value = 20 * log10(rms / 2e-5)

    def __add__(self, other):
        return Level(10*log10(10**(self.value/10) + 10**(other.value/10)))

    def __str__(self):
        return str(self.value)

    def __call__(self):
        return self.value

```

```python
l1 = Level(74.2)
l2 = Level(72.1)
l3 = l1 + l2
l3.value
```
