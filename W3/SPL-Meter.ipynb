## Ein Typisches SPL Meter

SPL = Schalldruckpegel = Sound Pressure Level


Zeitbewertung (Fast, Slow, Impulse)

Fast: 125 ms Integrationszeit - reagiert schnell, zeigt Spitzenwerte

Slow: 1 Sekunde Integrationszeit - geglättete, stabile Werte

Impulse: 35 ms Anstieg, 1,5 s Abfall - für kurze impulsive Ereignisse (Hammerschläge, Schüsse)

Frequenzbewertung (Z, A)

A-Bewertung (dBA): Angepasst an menschliches Gehör bei mittleren Lautstärken, weniger empfindlich bei tiefen/hohen Frequenzen

Z-Bewertung (dBZ): Linear/Flat - alle Frequenzen gleich stark, für technische Messungen

Oktavbandfilter und Terzbandfilter

Oktavbandfilter: Teilt Frequenzspektrum in Bänder, wobei jedes Band eine Oktave (Frequenzverdopplung) breit ist

Terzbandfilter: Feinere Auflösung - ein Drittel Oktave breite Bänder

C-Bewertung (dBC): 

Flacher als A-Bewertung, näher an linear - für hohe Pegel oder technische Messungen

CLI (Command Line Interface): Textbasierte Steuerung durch Befehle im Terminal. Schnell und automatisierbar, aber erfordert Kenntnis der Befehle.

TUI (Text User Interface): Textbasierte Oberfläche mit Menüs und Navigation per Tastatur. Intuitiver als CLI, aber ohne grafische Elemente.



Modelle
Pi Zero W: WiFi + Bluetooth
Pi Zero 2 W: Quad-Core, WiFi + Bluetooth
Pi Zero (ohne W): Kein WiFi/Bluetooth, günstiger
Bildschirm-Schnittstellen
1. Mini-HDMI (nur Pi Zero)
Standard: Mini-HDMI zu HDMI Adapter erforderlich
Auflösung: Bis 1080p
Verwendung: Normale HDMI-Monitore/TVs
2. DSI (Display Serial Interface)
Anschluss: 15-poliger Steckverbinder auf der Platine
Verwendung: Offizielle Raspberry Pi Touchscreens
Vorteil: Dediziert, keine CPU-Last
3. GPIO (SPI/I2C Displays)
Kleine Displays: 0.96"-3.5" TFT/OLED
Protokolle: SPI (schnell), I2C (einfach)
Beispiel: SSD1306 OLED, ST7735 TFT
Mikrofon-Anschlüsse
1. USB-Mikrofone
Einfachste Methode
Adapter: USB OTG Adapter erforderlich
Vorteil: Plug & Play, gute Qualität
2. I2S-Mikrofone
Direkt an GPIO: Pins 3, 5, 12, 35, 38, 40
Beispiel: INMP441, MEMS-Mikrofone
Vorteil: Digitale Qualität, keine USB-Last
3. USB-Soundkarte + Mikrofon
Kombination: USB-Soundkarte mit 3.5mm Mikrofon
Vorteil: Flexibel, verschiedene Mikrofone möglich
Touchscreen-Unterstützung
1. Offizieller Raspberry Pi Touchscreen
Größen: 7" (DSI), neuere Modelle
Touch: Kapazitiv, volle Unterstützung
Installation: DSI-Kabel + GPIO für Touch
2. USB-Touchscreens
Größen: 3.5"-10" verfügbar
Anschluss: USB OTG Adapter
Treiber: Meist automatisch erkannt
3. GPIO-Touch-Displays
Kombination: TFT + Touch Controller
Beispiel: ILI9341 mit XPT2046 Touch
Programmierung: Python-Bibliotheken verfügbar
Speicherkarten-Anforderungen
1. Geschwindigkeit
Minimum: Class 10 (10 MB/s)
Empfohlen: U3/V30 (30 MB/s Schreibrate)
Wichtig: Für OS und Programme
2. Kapazität
Minimum: 8 GB (Raspberry Pi OS Lite)
Empfohlen: 16-32 GB (mit Desktop)
Maximum: 2 TB (theoretisch)
3. Qualität
Marken: SanDisk, Samsung, Kingston
Typ: microSDHC (32GB) oder microSDXC (>32GB)
Besonderheit: "A1" oder "A2" Rating für bessere App-Performance
4. Wichtige Hinweise
Backup: Immer Backup der Karte machen
Formatierung: FAT32 für <32GB, exFAT für >32GB
Burning Tool: Raspberry Pi Imager verwenden
